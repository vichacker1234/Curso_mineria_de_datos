```python
#Se va a usar este cuaderno para el eejrcicio
#Enlace a la siguiente funci√≥n:
import math
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Se definen algunos par√°metros para sen(2x)
x=np.linspace(0,4*math.pi,100)
y=np.sin(2*x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel(r'$\sen(x)$')
plt.title("gr√°fica de sen(x)")
plt.show()
```
